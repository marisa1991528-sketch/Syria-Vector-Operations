データクリーニング：壊れたPolygonの検知・修正

使用技術：
Shapely is_valid
Shapely explain_validity
Shapely buffer(0)

対象データ：
サンプルPolygon（正常Geometry）
サンプルPolygon（自己交差Geometry）

分析手法：
・Geometry Validation
・Geometry Repair

その他：
・正常なPolygonと自己交差した(壊れた)Polygonを作成。
・is_valid によりGeometryの正常性を確認。
・explain_validity() によりGeometryエラーの原因を解析。
・Self-intersection（自己交差）を検知。
・buffer(0) によりGeometryを自動修復。
・修復後に再度 is_valid を実行し、正常Geometryであることを確認。

```python
# Core Libraries
# Geometry検証・修正に必要なライブラリを読み込む

from shapely.geometry import Polygon
from shapely.validation import explain_validity
```

```python
# ETL Extract
# 正常なPolygonと、自己交差した（壊れた）Polygonを作成する

valid_poly = Polygon([
    (0, 0),
    (0, 10),
    (10, 10),
    (10, 0),
    (0, 0)
])

broken_poly = Polygon([
    (0, 0),
    (10, 10),
    (0, 10),
    (10, 0),
    (0, 0)
])
```

```python
# Check
# 正常なPolygonと壊れたPolygonのvalidityを確認する

print(valid_poly.is_valid)
print(broken_poly.is_valid)
```

```python
# ETL Transform
# buffer(0) を使用して壊れたPolygonを修正する

fixed_poly = broken_poly.buffer(0)
```

```python
# Check
# 修正後のPolygonが正常Geometryになっているか確認する

print(fixed_poly.is_valid)
print(explain_validity(fixed_poly))
```
